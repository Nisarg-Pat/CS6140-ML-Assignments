## Solution 1:

### Step 1: Training the model: 

Calculating the probabilities $P(y = C_k)$ and $P(x_i=X_{ij}|y=C_k)$ assuming the features are conditionally independent.
Here, $y = $ buys\_computer and $x_i$ is the $i^{th}$ feature.

We have:
\begin{equation}
P(y = C_k) = \frac{\text{count of } y = C_k}{N}
\end{equation}

Thus,
\begin{align}
P(\text{y = yes}) &= \frac{9}{14}\\
P(\text{y = yes}) &= \frac{5}{14}
\end{align}

Now,
\begin{equation}
P(x_i=X_{ij}|y=C_k) = \frac{\text{count of both $x_i=X_{ij}$ and $y=C_k$}}{\text{count of } y = C_k}
\end{equation}

So for age, we have
\begin{align}
P(\text{age=youth}|\text{y=yes}) &= \frac{2}{9}\\
P(\text{age=youth}|\text{y=no}) &= \frac{3}{5}\\
P(\text{age=middle_aged}|\text{y=yes}) &= \frac{4}{9}\\
P(\text{age=middle_aged}|\text{y=no}) &= \frac{0}{5}\\
P(\text{age=senior}|\text{y=yes}) &= \frac{3}{9}\\
P(\text{age=senior}|\text{y=no}) &= \frac{2}{5}
\end{align}

Simmilarly for income, we have
\begin{align}
P(\text{income=low}|\text{y=yes}) &= \frac{3}{9}\\
P(\text{income=low}|\text{y=no}) &= \frac{1}{5}\\
P(\text{income=medium}|\text{y=yes}) &= \frac{4}{9}\\
P(\text{income=medium}|\text{y=no}) &= \frac{2}{5}\\
P(\text{income=high}|\text{y=yes}) &= \frac{2}{9}\\
P(\text{income=high}|\text{y=no}) &= \frac{2}{5}
\end{align}

And for student, we have
\begin{align}
P(\text{student=yes}|\text{y=yes}) &= \frac{6}{9}\\
P(\text{student=yes}|\text{y=no}) &= \frac{1}{5}\\
P(\text{student=no}|\text{y=yes}) &= \frac{3}{9}\\
P(\text{student=no}|\text{y=no}) &= \frac{4}{5}
\end{align}

And for credit_rating, we have
\begin{align}
P(\text{credit_rating=fair}|\text{y=yes}) &= \frac{6}{9}\\
P(\text{credit_rating=fair}|\text{y=no}) &= \frac{2}{5}\\
P(\text{credit_rating=excellent}|\text{y=yes}) &= \frac{3}{9}\\
P(\text{credit_rating=excellent}|\text{y=no}) &= \frac{3}{5}
\end{align}

### Step 2: Finding the posteriors:

We have:
\begin{align}
    P(y = C_k|x_1, x_2, ..., x_m) &\propto P(y = C_k, x_1, x_2, ..., x_m)\\
    P(y = C_k, x_1, x_2, ..., x_m) &= P(y = C_k) \prod_{i=1}^{m} P(x_i|y=C_k)
\end{align}

Thus, we have:

$P(\text{y = yes, age = youth, income = medium, student = yes, credit_rating = fair})$
$=P(\text{age=youth}|\text{y=yes})*P(\text{income=medium}|\text{y=yes})*P(\text{student=yes}|\text{y=yes})*P(\text{credit_rating=fair}|\text{y=yes})*P(\text{y = yes})$
\begin{align}
    &=\frac{2}{9}*\frac{4}{9}*\frac{6}{9}*\frac{6}{9}*\frac{9}{14}\\
    &=0.0282
\end{align}

And, we have:
$P(\text{y = no, age = youth, income = medium, student = yes, credit_rating = fair})$
$=P(\text{age=youth}|\text{y=no})*P(\text{income=medium}|\text{y=no})*P(\text{student=yes}|\text{y=no})*P(\text{credit_rating=fair}|\text{y=no})*P(\text{y = no})$
\begin{align}
    &=\frac{3}{5}*\frac{2}{5}*\frac{1}{5}*\frac{2}{5}*\frac{5}{14}\\
    &=0.0069
\end{align}

### Step: 3 Predicting the outcome based on the posteriors:

Since $P(\text{y = yes, age = youth, income = medium, student = yes, credit_rating = fair}) > P(\text{y = no, age = youth, income = medium, student = yes, credit_rating = fair})$

We have $P(\text{y = yes}| \text{age = youth, income = medium, student = yes, credit_rating = fair}) > P(\text{y = no}| \text{age = youth, income = medium, student = yes, credit_rating = fair})$

And hence, $y^* = $ yes.

### Thus Naive Bayes predicts that the customer will buy a computer.

## Solution 2:

### Part 1:

\begin{align}
    \text{False Positives} &= \text{Number of negative samples incorrectly predicted as positive}\\
    &=3
\end{align}

\begin{align}
    \text{False Negative} &= \text{Number of positive samples incorrectly predicted as negative}\\
    &=1
\end{align}

### Part 2:

We have:
\begin{align}
    \text{Accuracy} &= \frac{\text{True Positive}+\text{True Negative}}{\text{Total Samples}}\\
                    &= \frac{33+72}{33+72+3+1}\\
                    &= \frac{105}{109}\\
                    &=0.9633
\end{align}
#### Thus the accuracy of the test is 96.33%.


### Part 3:

We have:
\begin{align}
    \text{Precision} &= \frac{\text{True Positive}}{\text{True Positive}+\text{False Positive}}\\
                    &= \frac{33}{33+3}\\
                    &= \frac{33}{36}\\
                    &=0.9167
\end{align}

\begin{align}
    \text{Recall} &= \frac{\text{True Positive}}{\text{True Positive}+\text{False Negative}}\\
                    &= \frac{33}{33+1}\\
                    &= \frac{33}{34}\\
                    &=0.9706
\end{align}

#### Thus the precision of the test is 91.67% and the recall is 97.06%.

### Part 4:

We have:
\begin{align}
    \text{F1 score} &= \frac{2}{\frac{1}{Precision}+\frac{1}{Recall}}\\
                    &= \frac{2}{\frac{1}{0.9167}+\frac{1}{0.9706}}\\
                    &=0.9429
\end{align}

#### Thus the F-1 score of the test is 0.9429.

````python
## Read the dataset
import numpy as np
import pandas
from sklearn.model_selection import train_test_split
````

````python
def load_data(data_dir):
    """
    :param data_dir: Directory of Data file
    :return: features and labels
        data: input features
        labels: output features
    """

    dataframe = pandas.read_excel(data_dir, sheet_name=0)
    data = dataframe.drop("ID", axis=1).drop("default payment next month", axis=1).values
    labels = dataframe["default payment next month"].values
    return data, labels


def normalize_features(data):
    col_min = np.min(data, axis=0)
    col_max = np.max(data, axis=0)
    data_n = (data - col_min) / (col_max - col_min)
    return data_n
````


### 2)  Implement logistic regression using batch gradient descent and evaluation (20 points)
Algorithm can be given as follows:

```for j in 0 -> max_iteration: 
    for i in 0 -> m: 
        theta += (alpha / m) * (y[i] - h(x[i])) * x_bar
```

````python
class LogisticRegression:

    def __init__(self):
        self.params = None
        return

    def sigmoid(self, z):
        return 1.0 / (1.0 + np.exp(-z))

    def compute_cost(self, ip, op):
        """
        Cost function in linear regression where the cost is calculated
        ip: input variables
        op: output variables
        params: corresponding parameters
        Returns cost
        """
        num_samples = len(ip)
        cost_sum = 0.0
        for x, y in zip(ip, op):
            y_hat = self.sigmoid(np.dot(self.params, x))
            cost_sum = cost_sum - ((y * np.log(y_hat)) + ((1 - y) * (np.log(1 - y_hat))))

        cost = cost_sum / num_samples

        return cost

    def batchify(self, ip, op, batch_size):
        X_Batch = []
        Y_Batch = []

        for i in range(int(np.ceil(len(ip) / batch_size))):
            X_Batch.append([])
            Y_Batch.append([])
        for i in range(len(ip)):
            X_Batch[int(i / batch_size)].append(ip[i])
            Y_Batch[int(i / batch_size)].append(op[i])
        num_batches = len(X_Batch)
        return X_Batch, Y_Batch, num_batches

    def batchGradientDescent(self, ip, op, paramss, alpha, max_iter, batch_size=1):
        """
        Compute the params for logistic regression using batch gradient descent
        ip: input variables
        op: output variables
        params: corresponding parameters
        alpha: learning rate
        max_iter: maximum number of iterations
        Returns parameters, cost
        """
        # initialize iteration, number of samples, cost and parameter array
        iteration = 0
        num_samples = len(ip)
        num_params = len(paramss)
        self.params = paramss

        # batchify the data into mini-batches
        X_Batch, Y_Batch, num_batches = self.batchify(ip, op, batch_size)
        cost = np.zeros(max_iter)

        # Compute the cost and store the params for the corresponding cost
        while iteration < max_iter:
            cost[iteration] = self.compute_cost(ip, op)

            print('--------------------------')
            print(f'iteration: {iteration}')
            print(f'cost: {cost[iteration]}')

            for k in range(num_batches):
                # Apply batch gradient descent
                batch_length = len(X_Batch[k])
                partialDifferentials = np.zeros(num_params)
                for i in range(batch_length):
                    sgm = self.sigmoid(np.dot(self.params, X_Batch[k][i]))
                    for j in range(num_params):
                        partialDifferentials[j] += (sgm - Y_Batch[k][i]) * X_Batch[k][i][j]
                self.params = self.params - (alpha / batch_length * partialDifferentials)
            iteration += 1

        return self.params, cost

    def predictOne(self, x):
        sgm = self.sigmoid(np.dot(self.params, x))
        if sgm >= 0.5:
            return 1
        else:
            return 0

    def predict(self, x):
        y_cap = [self.predictOne(x[i]) for i in range(len(x))]
        return y_cap

    def calc_accuracy(self, ip, op):
        count = 0
        for i in range(len(ip)):
            if self.predictOne(ip[i]) == op[i]:
                count += 1
        return count / len(ip)

    def evaluate(self, x, y):
        TP, FP, TN, FN = 0, 0, 0, 0
        for i in range(len(x)):
            y_cap = self.predictOne(x[i])
            if y[i] == 1 and y_cap == 1:
                TP += 1
            elif y[i] == 1 and y_cap == 0:
                FN += 1
            elif y[i] == 0 and y_cap == 0:
                TN += 1
            elif y[i] == 0 and y_cap == 1:
                FP += 1
        print(f"True Positive: {TP}")
        print(f"False Positive: {FP}")
        print(f"True Negative: {TN}")
        print(f"False Negative: {FN}")
        print(f"Accuracy: {(TP+TN)/(TP+FP+TN+FN)}")
        if TP+FP == 0:
            print("Precision cannot be calculated as TP+FP = 0")
        else:
            print(f"Precision: {TP/(TP+FP)}")
        if TP+FN == 0:
            print("Recall cannot be calculated as TP+FN = 0")
        else:
            print(f"Recall: {(TP/(TP+FN))}")
````

### 3) Implementation of perceptron.(20 points) 

````python
class Perceptron:
    # constructor
    def __init__(self):
        self.eta = 0
        self.num_epochs = 0
        self.x = None
        self.y = None
        self.num_ip = 0
        self.w = None

    # model
    def model(self, x, y, eta=1.0, num_epochs=100):
        self.num_ip = len(x)
        self.x = x
        self.y = [1 if y[i] == 1 else -1 for i in range(self.num_ip)]
        self.w = np.zeros(self.x.shape[1])
        self.eta = eta
        self.num_epochs = num_epochs

    # fitting the model
    def fit(self):
        epoch = 0
        while epoch < self.num_epochs:
            print(f'Epoch {epoch}')
            for i in range(self.num_ip):
                y_pred = self.predictOne(self.x[i])
                if y_pred * self.y[i] <= 0:
                    self.w = self.w + (self.eta * self.y[i] * self.x[i])
            epoch += 1

    # predictor to predict on the data based on w
    def predictOne(self, x):
        val = np.dot(self.w, x)
        if val > 0:
            return 1
        elif val == 0:
            return 0
        else:
            return -1

    def predict(self, x):
        y_cap = [self.predictOne(x[i]) for i in range(len(x))]
        return y_cap

    def calc_accuracy(self, x, y):
        y_pred = [self.predictOne(x[i]) for i in range(len(x))]
        count = 0
        for i in range(len(x)):
            if y_pred[i] == 1 and y[i] == 1:
                count += 1
            elif y_pred[i] == -1 and y[i] == 0:
                count += 1
        return count / len(x)

    def evaluate(self, x, y):
        TP, FP, TN, FN = 0, 0, 0, 0
        for i in range(len(x)):
            y_cap = self.predictOne(x[i])
            if y[i] == 1 and y_cap == 1:
                TP += 1
            elif y[i] == 1 and y_cap == -1:
                FN += 1
            elif y[i] == 0 and y_cap == -1:
                TN += 1
            elif y[i] == 0 and y_cap == 1:
                FP += 1
            else:
                print(f'{i}: {y[i]}, {y_cap}')
        print(f"True Positive: {TP}")
        print(f"False Positive: {FP}")
        print(f"True Negative: {TN}")
        print(f"False Negative: {FN}")
        print(f"Accuracy: {(TP + TN) / (TP + FP + TN + FN)}")
        if TP + FP == 0:
            print("Precision cannot be calculated as TP+FP = 0")
        else:
            print(f"Precision: {TP / (TP + FP)}")
        if TP + FN == 0:
            print("Recall cannot be calculated as TP+FN = 0")
        else:
            print(f"Recall: {(TP / (TP + FN))}")
````

### 4) Apply 80-20 split on data to prepare training and test sets. Report training and test results in terms of accuracy, precision and recall for both logistic regression and perceptron. (20)

````python
# Sample training code cell change according to your variables and structure

# Training the model
#reserve the test data, do not use them for cross-validation!


data, labels = load_data("Credit card Default.xlsx")
data_N = normalize_features(data)

x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.20)
x_train = np.append(np.ones((x_train.shape[0],1)),x_train, axis=1)
x_test = np.append(np.ones((x_test.shape[0],1)),x_test, axis=1)

x_train_N, x_test_N, y_train_N, y_test_N = train_test_split(data_N, labels, test_size=0.20)
x_train_N = np.append(np.ones((x_train_N.shape[0],1)),x_train_N, axis=1)
x_test_N = np.append(np.ones((x_test_N.shape[0],1)),x_test_N, axis=1)
````

````python
#Logistic Regression for raw data
intitial_params = np.ones(x_train.shape[1])
learning_rate = 0.01
max_iter = 200
batch_size = 64

logisticRegressionRaw = LogisticRegression()
logisticRegressionRaw.batchGradientDescent(x_train, y_train, intitial_params, learning_rate,max_iter, batch_size)
````

````python
#Logistic Regression for normalized data
intitial_params = np.ones(x_train_N.shape[1])
learning_rate = 0.01
max_iter = 200
batch_size = 64

logisticRegressionNormalized = LogisticRegression()
logisticRegressionNormalized.batchGradientDescent(x_train_N, y_train_N, intitial_params, learning_rate,max_iter, batch_size)
````

````python
#Perceptron for raw data
perceptronRaw = Perceptron()
perceptronRaw.model(x_train, y_train, eta=0.01, num_epochs=1000)
perceptronRaw.fit()
````

````python
#Perceptron for normalized data
perceptronNormalized = Perceptron()
perceptronNormalized.model(x_train_N, y_train_N, eta=0.01, num_epochs=1000)
perceptronNormalized.fit()
````
